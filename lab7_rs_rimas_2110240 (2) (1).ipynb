```python
#1
class HybridRecommender:
    MODEL_NAME = 'Hybrid'

    def __init__(self, cb_rec_model, cf_rec_model, items_df, cb_ensemble_weight=1.0, cf_ensemble_weight=1.0, cascade=False):
        self.cb_rec_model = cb_rec_model
        self.cf_rec_model = cf_rec_model
        self.cb_ensemble_weight = cb_ensemble_weight
        self.cf_ensemble_weight = cf_ensemble_weight
        self.items_df = items_df
        self.cascade = cascade

    def get_model_name(self):
        return self.MODEL_NAME

    def recommend_items(self, user_id, items_to_ignore=[], topn=10, verbose=False, cascade_topn=None):
        # Default cascade_topn to topn if not provided
        if cascade_topn is None:
            cascade_topn = topn

        # Get top-1000 Content-based recommendations
        cb_recs_df = self.cb_rec_model.recommend_items(
            user_id, items_to_ignore=items_to_ignore, verbose=verbose, topn=1000
        ).rename(columns={'recStrength': 'recStrengthCB'})

        # Get top-1000 Collaborative filtering recommendations
        cf_recs_df = self.cf_rec_model.recommend_items(
            user_id, items_to_ignore=items_to_ignore, verbose=verbose, topn=1000
        ).rename(columns={'recStrength': 'recStrengthCF'})

        if self.cascade:
            # Cascading pipeline hybridization
            cb_topn_ids = cb_recs_df.head(cascade_topn)['contentId']
            cf_recs_filtered = cf_recs_df[cf_recs_df['contentId'].isin(cb_topn_ids)].copy()

            # Supplement with additional CF recommendations if fewer than `topn` items are found
            if len(cf_recs_filtered) < topn:
                additional_items = cf_recs_df[~cf_recs_df['contentId'].isin(cb_topn_ids)].head(topn - len(cf_recs_filtered))
                cf_recs_filtered = pd.concat([cf_recs_filtered, additional_items])

            cf_recs_filtered['recStrengthHybrid'] = cf_recs_filtered['recStrengthCF']
            recommendations_df = cf_recs_filtered.sort_values('recStrengthHybrid', ascending=False).head(topn)
        else:
            # Parallelized hybridization
            recs_df = cb_recs_df.merge(cf_recs_df, how='outer', on='contentId').fillna(0.0)
            recs_df['recStrengthHybrid'] = (
                recs_df['recStrengthCB'] * self.cb_ensemble_weight
                + recs_df['recStrengthCF'] * self.cf_ensemble_weight
            )
            recommendations_df = recs_df.sort_values('recStrengthHybrid', ascending=False).head(topn)

        if verbose:
            if self.items_df is None:
                raise Exception('"items_df" is required in verbose mode')

            recommendations_df = recommendations_df.merge(
                self.items_df, how='left', on='contentId'
            )[['recStrengthHybrid', 'contentId', 'title', 'url', 'lang']]

        return recommendations_df
```

Why Are the Results Different?
The difference arises because the input filtering set (top 10 vs. top 20) impacts how many items from collaborative filtering are retained:

Top-10 Cascading:
Smaller, more focused filtering set.
Produces highly specific recommendations, as only items in the top 10 from the content-based model are considered.
Top-20 Cascading:
Broader filtering set.
Increases the chance of diverse items passing through, leading to more varied recommendations.
When to Use Each Approach?
Top-10 Cascading:

Suitable when you want focused, high-quality recommendations.
Reduces noise by narrowing down collaborative filtering results to match the user's strongest preferences.
Top-20 Cascading:

Suitable when you want diverse recommendations.
Allows for exploration beyond the most strongly matched items.
